# Automação de Sistemas e Processos com Python

### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
# INSTALAR PIP INSTALL PYAUTOGUI
# PIP INSTALL PYPERCLIP
```

```python
#!pip install pyautogui
#!pip install pyperclip
```

```python
import pyautogui
import pyperclip
import time
# PASSO 1 => ENTRAR NO SISTEMA DA EMPRESA (NO CASO VAI SER O LINK DO DRIVER
pyautogui.PAUSE = 1
#pyautogui.hotkey("ctrl", "t")  
pyautogui.hotkey("command", "t")
pyperclip.copy("https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing")
pyautogui.hotkey("command", "v")    
pyautogui.hotkey("enter")   

#DEMORA ALGUNS SEGUNDOS
time.sleep(5)

# PASSO 2 => NAVEGAR NO SISTEMA E ENCONTRAR A BASE DE DADOS 
pyautogui.click(x=1730, y=358, clicks=2, button='left')
pyautogui.hotkey("enter")
time.sleep(2)

# PASSO 3 => EXPORTAR/FAZER DOWNLOAD DA BASE DE DADOS
pyautogui.click(x=1776, y=421) #clicar no arquivo
pyautogui.click(x=3078, y=247) #clicar nos 3 pontinhos
pyautogui.click(x=2803, y=659) #clicar no fazer download
time.sleep(5)
pyautogui.click(x=2803, y=659) #clicar no fazer download
```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
# PASSO 4 => IMPORTAR A BASE DE DADOS PARA PYTHON
import pandas as pd
tabela = pd.read_excel(r"/Users/joaodematejr/Downloads/Vendas - Dez.xlsx")
display(tabela)
```

```python
# PASSO 5 => CALCULAR OS INDICADORES
# FATURAMENTO = SOMA DA COLUNA VALOR FINAL
faturamento = tabela['Valor Final'].sum()
# QUANTIDADE DE PRODUTOS - SOMA DA COLUNA QUANTIDADE
quantidade = tabela['Quantidade'].sum()

display(quantidade)
display(faturamento)
```

### Vamos agora enviar um e-mail pelo gmail

```python
# PASSO 6 => ENVIAR UM EMAIL PARA DIRETORIA COM RELATÓRIO
# ABRIR O EMAIL https://mail.google.com/mail/u/0/
pyautogui.hotkey("command", "t")
pyperclip.copy("https://mail.google.com/mail/u/0/")
pyautogui.hotkey("command", "v")    
pyautogui.press("enter")  
time.sleep(7)
#CLICAR NO ESCREVER
pyautogui.click(x=1422, y=262)
time.sleep(2)
# ESCREVE O EMAIL DESTINARIO
pyautogui.write("joaodematejr@gmail.com")
pyautogui.press("tab") #SELECIONA O DESTINATARIO
pyautogui.press("tab") # PASSA PARA O CAMPO DE ASSUNTO

# ESCREVE O ASSUNTO
pyperclip.copy("Relatório de Vendas")
pyautogui.hotkey("command", "v")   
pyautogui.press("tab") #SELECIONA O DESTINATARIO
# ESCREVE O CORPO DO EMAIL
texto = f"""
Prezados, bom dia

O faturamento de ontem foi de: R${faturamento:,.2f}
A quantidade de produtos foi de: {quantidade:,}

Abç
João """
pyperclip.copy(texto)
pyautogui.hotkey("command", "v")   
# ENVIA O EMAIL
pyautogui.hotkey("command", "enter") 
```
